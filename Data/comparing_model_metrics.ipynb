```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/claireleblanc/Downloads/Model_Results - Sheet1.csv")
data
```

```python
import numpy as np

plt.scatter(data[(data["Negative K1?"] == "No") & (data["Negative K2?"] == "No")]["R^2 abundance"], data[(data["Negative K1?"] == "No") & (data["Negative K2?"] == "No")]["R^2 activity"],label="Positive K1 and K2 predicted values")
plt.scatter(data[(data["Negative K1?"] != "No") | (data["Negative K2?"] != "No")]["R^2 abundance"], data[(data["Negative K1?"] != "No") | (data["Negative K2?"] != "No")]["R^2 activity"],label="Negative K1 or K2 predicted values")

plt.scatter(data[(data["R^2 activity"] == 0.768)]["R^2 abundance"], data[((data["R^2 activity"] == 0.768))]["R^2 activity"],label="Model 1")
plt.scatter(data[(data["R^2 activity"] == 0.76)]["R^2 abundance"], data[((data["R^2 activity"] == 0.76))]["R^2 activity"],label="Model 2")
plt.scatter(data[(data["R^2 activity"] == 0.751)]["R^2 abundance"], data[((data["R^2 activity"] == 0.751))]["R^2 activity"],label="Model 3")
plt.scatter(data[(data["R^2 activity"] == 0.718) & (data["R^2 abundance"] == 0.296)]["R^2 abundance"], data[((data["R^2 activity"] == 0.718) & (data["R^2 abundance"] == 0.296))]["R^2 activity"],label="Model 4")
plt.scatter(data[(data["R^2 activity"] == 0.707) & (data["R^2 abundance"] == 0.298)]["R^2 abundance"], data[((data["R^2 activity"] == 0.707) & (data["R^2 abundance"] == 0.298))]["R^2 activity"],label="Model 4")


plt.xlabel("Abundance R^2")
plt.ylabel("Activity R^2")
plt.legend()
```

```python
plt.scatter(data[(data["Bound -> Activity"] == "Linear") & (data["Bound -> Activity"] == "Linear")]["R^2 abundance"], data[(data["Bound -> Activity"] == "Linear") & (data["Bound -> Activity"] == "Linear")]["R^2 activity"],label="Linear activation function")
plt.scatter(data[(data["Bound -> Activity"] != "Linear") | (data["Bound -> Activity"] != "Linear")]["R^2 abundance"], data[(data["Bound -> Activity"] != "Linear") | (data["Bound -> Activity"] != "Linear")]["R^2 activity"], label = "Hill Activation Function")
plt.xlabel("Abundance R^2")
plt.ylabel("Activity R^2")
plt.legend()
```
